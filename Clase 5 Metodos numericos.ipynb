## Métodos númericos

Ahora, veremos como se usa scipy.integrate, veremos opciones y metodos que ofrece.

Ejemplo, cuarto de circulo:

$$ I = \int_0^1 \sqrt{1-x^2} dx = \frac{\pi}{4} $$

En mis tiempos lo haciamos con la libreria quadpack de fortran, python recrea el mismo metodo.

```python
# supongamos 10 intervalos e integracion de Riemann
import numpy as np

funcion = lambda x: np.sqrt(1-x*x)

teorico = np.pi/4
print("Resultado teorico =", teorico) #ya hay un error de redondeo
```

```python
from scipy.integrate import quad
import time

inicio = time.time()
res = quad(funcion, 0, 1)
fin = time.time()
print("Tiempo calculo en (s)", fin-inicio)
res # el primer elemento de la tupla es el resultado, el segundo es una cota superior del error
```

Además de medir el error, en general debemos considerar el tiempo que demora los calculos, evidentemente en procesos o transacciones. Revise magnitud de errores, tiempo de ejecución, que pasa cuando crece el numero de intervalos de la clase anterior con el tiempo de ejecución.

```python
# Algo un poco mejor... aunque si quieren no lo usen

inicio = time.time()
res_lambda = quad(lambda x: np.sqrt(1-x*x), 0, 1)
fin = time.time()
print("Tiempo calculo en (s)", fin-inicio)
res_lambda # es lo mismo.... discutir pq demora mas, ahora es el tiempo completo...
```

Como hacemos los intervalos en nuestra programación....  sin embargo scipy.integrate puede.

```python
#revisemos el ejemplo del video....

inicio = time.time()
res = quad(lambda x: 1/(1+x*x), -np.inf, np.inf)
fin = time.time()
print("Tiempo calculo en (s)", fin-inicio)
print(res)
```
