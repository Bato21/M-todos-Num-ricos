### Problema: período no lineal del péndulo

El periodo de oscilación de un péndulo simple (longitud $L$, gravedad $g$, sin considerar el roce) para una amplitud angular máxima $\alpha_0$ se obtiene por la integral elíptica:

$$T(\alpha_0) = 4 \sqrt{\frac{L}{g}} \int_{0}^{\frac{\pi}{2}} \frac{d\theta}{\sqrt{1-sen^2(\frac{\alpha_0}{2}) sen^2(\theta)}}$$

La aproximación de un ángulo pequeño (probablemente visto en física), es:

$$ T \approx 2 \pi \sqrt{\frac{L}{g}}$$

```python
# Bonus, tambien hay integracion simbolica.... veamos nuestro ejemplo

import sympy

x = sympy.Symbol('x')
eliptica = sympy.Integral(1/sympy.sqrt((1-sympy.sin(x)**2/5)), (x, 0, sympy.pi/4))
eliptica
```

```python
eliptica.doit()
```

```python
print(eliptica.evalf())
```

```python
# otra forma
circ2 = sympy.integrate(1/sympy.sqrt((1-sympy.sin(x)**2/5)), (x, 0, sympy.pi/4))
circ2
```

```python
# otra forma
circ3 = sympy.integrate(1/sympy.sqrt((1-sympy.sin(x)**2/5)), x)
circ3
```
