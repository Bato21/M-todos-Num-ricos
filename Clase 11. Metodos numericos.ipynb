### Metodo de Newton-Raphson

https://es.wikipedia.org/wiki/M%C3%A9todo_de_Newton

Se explica bien en https://www.youtube.com/watch?v=tX9ecFstUUk

Hay que tener cuidado si la derivada da 0, en mi ejemplo xi=0. Discutir, ideas para mejorar (mover un poquito el xi?)

```python
xi = 1
funcion = lambda x: x*x-2
df = lambda x: 2*x
maxiter = 1000 # como no sabemos si converge, debemos protegernos para terminar
tolerancia = 0.000001
iteraciones = 0

while iteraciones < maxiter:
    xnew = xi - funcion(xi)/df(xi)
    if abs(xnew-xi) < tolerancia:
        break
    xi = xnew
    iteraciones += 1

print(2**.5, xi, iteraciones)
# ustedes deberian calcular error y tiempos de ejecucion
# se puede hacer una funcion con el ciclo while y pasar f y df como parametro
```

### Metodo de la secante

Se parece a la forma de calculo de la falsa posicion pero sin la derivada, es como lo hicimos con el metodo de Euler donde la derivada es la diferencia en ambos ejes. 

https://es.wikipedia.org/wiki/M%C3%A9todo_de_la_secante

Se explica en: https://www.youtube.com/watch?v=LDYpMHsrv_c

```python
x0 = 0
x1 = 1
funcion = lambda x: x*x-2
maxiter = 1000 # como no sabemos si converge, debemos protegernos para terminar
tolerancia = 0.000001
iteraciones = 0
while iteraciones < maxiter:
    fx0 = funcion(x0)
    fx1 = funcion(x1)
    xnew = x1 - fx1 * (x1 - x0) / (fx1 - fx0)
    if abs(xnew-x1) < tolerancia:
        break
    x1,x0 = xnew,x1
    iteraciones += 1

print(2**.5, xi, iteraciones)
# ustedes deberian calcular error y tiempos de ejecucion
# se puede hacer una funcion con el ciclo while y pasar f como parametro
# se pareceria mas al metodo anterior si x0 es mas parecido a x1.... un poco mas o menos... revisar
```
