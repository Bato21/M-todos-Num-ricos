## Bitwise operators

Los bitwise operators son operadores que realizan operaciones lógicas sobre cada bit en memoria. Son muy eficientes ya que pueden ser enviadas directamente a las compuertas lógicas que componen un procesador.

Aprende usted que es una compuerta logica en: https://en.wikipedia.org/wiki/Logic_gate (hay traducción al español disponible)

### AND ( & ) 
Fijese que es un solo caracter y difiere de la comparador and de expresiones logicas

```python
a = 123
b = 456
print("  " + bin(a) + "\n" + bin(b))
```

```python
c = a&b
print(bin(c), c)
```

### OR ( | )

```python
d = a|b
print(bin(d), d)
```

### XOR ( ^ )

```python
e = a^b
print(bin(e), e)
```

### Left shift ( << )

```python
f = a << 3
print(bin(a), bin(f)) # agrega 3 posiciones al final
```

### Right shift ( >> )

```python
g = a >> 2
print(bin(a), bin(g)) # quita 2 posiciones al final
```

## Desafio

Se podra sumar o multiplicar (numeros positivos), con los operadores vistos arriba... si.. explicacion en pizarra

```python
def suma(n1, n2):
    if n2 == 0:
        return n1
    resultado = n1^n2
    reserva = (n1&n2) << 1
    return suma(resultado, reserva)
```

```python
# probaremos la funcion recursiva suma aleatoriamente
import random

try:
    for _ in range(100000): # cien mil experimentos
        a = random.randint(1000, 100000)
        b = random.randint(1000, 100000)
        assert (a+b) == suma(a,b)
except AssertionError:
    print("Fallo en suma de", a, "+", b)
else:
    print("Funciono bien")
```

```python
def multiplicacion(n1, n2):
    resultado = 0
    while n2 > 0:
        if n2&1 == 1:
            resultado = suma(resultado, n1)
        n1 <<= 1
        n2 >>= 1
    return resultado
```

```python
# probaremos la funcion iterativa multiplicacion aleatoriamente
import random

try:
    for _ in range(100000): # cien mil experimentos
        a = random.randint(1000, 10000)
        b = random.randint(1000, 10000)
        assert (a*b) == multiplicacion(a,b)
except AssertionError:
    print("Fallo en mult de", a, "*", b)
else:
    print("Funciono bien")
```
